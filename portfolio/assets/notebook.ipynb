This "MCTS" function starts by evaluating if the current state of the game already has a node representing it. If not, it creates one. Then it runs 100 times the simulated game from the first node, just to "get the search tree going". After that, it chooses the best node to expand and simulates a game from that node x times while the time is under 7 seconds. In the end, it evaluates all the children from the current state of the game and chooses the best play, using a score of wins/iteractions for each children.

```python
    def existing_node (self, matriz):
            for i in range (len(self.search_tree)):
                if np.array_equal(self.search_tree[i].matriz, matriz):
                    return self.search_tree[i]
            return False
```

```python
    def backpropagation (self, first_node, win):
            if win == 0:
                node = first_node
                while node != None:
                    node.ite += 1
                    node = node.father
            if win == 1:
                node = first_node
                while node != None:
                    node.ite += 1
                    node.wins += 1
                    node = node.father
```

This function does the backpropagation of iteractions and wins (if the case) from the leaf node (first_node) until the root.

```python
    def search_node (self):
            best_score = 0
            best_node = None
            for i in range (len(self.search_tree)):
                score = 0
                node = self.search_tree [i]
                if (node.father == None):
                    if (node.ite == 0):
                        score = 2
                    else:
                        score = node.wins/node.ite
                else:
                    if (node.ite == 0):
                        score = 2
                    else:
                        score = node.wins/node.ite + math.sqrt(2)*math.sqrt(2*(math.log(node.father.ite)/node.ite))
                if (score >= best_score):
                    best_score = score
                    best_node = node
            return best_node
```
